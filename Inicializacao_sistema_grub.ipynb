### **Inicialização do Sistema – Bootloaders – GRUB, LILO**

**Processos de Boot**


 *Etapa* | *Descrição* |
| --------- | --------------- |
| 1. Computador é ligado | Início do processo de boot |
| 2. BIOS | Inicializa o hardware e carrega o bootloader do MBR |
| 3. Bootloader (Estágio 1) | Carregado do MBR, responsável por carregar o bootloader do estágio 2 |
| 4. Bootloader (Estágio 2) | GRUB, LILO, etc. Carrega o kernel do sistema operacional |
| 5. Kernel | Carregado pelo bootloader, responsável por inicializar o sistema operacional |
| 6. Init | Processo que inicializa os demais processos do sistema operacional |
| 7. Serviços e Aplicativos | Inicializados pelo Init, fornecem funcionalidades ao sistema |
   

**MBR**

    • Master Boot Record, ou setor de partições, é um setor
    de boot de 512 bytes que é o primeiro setor de um disco particionado.

Funções:
      
    • Abrigar a tabela de partições primárias;
    • Inicializar sistemas operacionais a partir de um bootloader

**Estrutura do Master Boot Record (MBR)**



 Offset (Hex) | Offset (Dec) | Tamanho (Bytes) | Descrição |
| ------------ | ------------ | --------------- | ----------- |
| 0000h | 0 | 446 | Código do Boot Loader |
| 01BEh | 446 | 16 | Entrada da Tabela de Partições 1 |
| 01CEh | 462 | 16 | Entrada da Tabela de Partições 2 |
| 01DEh | 478 | 16 | Entrada da Tabela de Partições 3 |
| 01EEh | 494 | 16 | Entrada da Tabela de Partições 4 |
| 01FEh | 510 | 2 | Assinatura do MBR |
