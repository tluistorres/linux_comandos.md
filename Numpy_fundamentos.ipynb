```python
import numpy as np

# Crie um array de exemplo
a1 = np.array([1, 2, 3, 4, 5])

# Calcule estatísticas
media = np.mean(a1)
mediana = np.median(a1)
desvio_padrao = np.std(a1)
minimo = np.min(a1)
maximo = np.max(a1)

print("Média:", media)
print("Mediana:", mediana)
print("Desvio Padrão:", desvio_padrao)
print("Mínimo:", minimo)
print("Máximo:", maximo)
```
